Q1.

*   Read the file into the Pandas data frame
*   Stock tickers have either “IN”, “IS”, or “IB” as suffixes with space between the ticker
and suffix. Remove the suffixes and the space from the tickers

Ans: We are using the re module is Python's built-in library which allows for pattern-based string manipulation
we are defining a function so that we can later use the python map function and apply it to the entire dataframe.
We could have done it in parts but this method is faster and more efficient.
Isintance is being used because it was throwing not a string error otherwise so to makesure we used Isinstance.

```python
import re
# Defining a function to remove the suffixes and spaces from the tickers
def remove_suffix(string):
    if isinstance(string, str):
        # Remove suffixes and extra spaces
        string = re.sub(r' (IN|IS|IB)$', '', string).replace(" ", "")
    return string


# Applying the function to the entire DataFrame and handling not string objects
df = df.apply(lambda col: col.map(remove_suffix) if col.dtype == 'object' else col)
df.head(100)
```

Mathematical Overview


---


Newton's Method for Square Roots:

Newton's method is an iterative technique for finding successively better approximations of the roots (or zeroes) of a real-valued function.
This method converges quadratically, meaning that the number of correct digits roughly doubles with each iteration.


---


Halley's Method for Square Roots:

Halley's method is a third-order iterative method for finding roots of functions. It can be considered an extension of Newton's method, using additional derivative information to achieve faster convergence.

This method converges cubically, meaning the number of correct digits triples with each iteration, leading to potentially faster convergence than Newton's method, especially when very high precision is required.


---


Karatsuba Algorithm for Square Roots:

The Karatsuba algorithm is primarily a fast multiplication method used for large integers. When adapted for square roots, it provides a way to handle large numbers efficiently by breaking the problem into smaller parts recursively.
The karatsuba_sqrt function essentially uses a method similar to Newton's method but relies on integer division and multiplication, which can be optimized for very large numbers due to the Karatsuba multiplication technique.


---


Performance Comparison

Halley's Method:

Advantage: Halley's method often has better execution time when high precision is required because it converges more quickly (cubic convergence) compared to Newton's method (quadratic convergence). This means fewer iterations are needed to reach a certain level of accuracy, making it faster in scenarios where precision is critical.
Disadvantage: The complexity of each iteration in Halley's method is higher due to the use of additional derivative calculations, which might make it slower if fewer iterations are needed or for problems requiring low precision.

Halley's Method excels in scenarios requiring high precision due to its faster convergence.


---



Karatsuba Square Root:

Advantage: The Karatsuba algorithm is particularly efficient for large integers because it reduces the complexity of multiplication. When applied to square roots, it can handle very large numbers effectively, making it faster for cases where the input size is extremely large.
Disadvantage: For smaller numbers or when the overhead of recursive calls outweighs the benefits of faster multiplication, Karatsuba may not be as efficient as Halley's or even Newton's method.

Karatsuba Square Root performs better with very large numbers because it optimizes the multiplication steps involved in iterative root-finding methods.


---



```python
import math
# Newton's Method for Square Root
def sqrt_newton(a, tolerance=1e-10, max_iterations=1000):
    if a < 0:
        raise ValueError("Cannot compute square root of a negative number")
    if a == 0:
        return 0
    x = a
    for _ in range(max_iterations):
        x_next = 0.5 * (x + a / x)
        if abs(x_next - x) < tolerance:
            return x_next
        x = x_next
    raise ValueError("Newton's method did not converge")

def sqrt_halley(a, tolerance=1e-10, max_iterations=1000):
    """Compute the square root of a number using Halley's method."""
    if a < 0:
        raise ValueError("Cannot compute square root of a negative number")

    if a == 0:
        return 0

    # Initial guess for the square root
    x = a
    for _ in range(max_iterations):
        f_x = x**2 - a
        f_prime_x = 2 * x
        f_double_prime_x = 2

        # Update the guess using Halley's method formula
        denominator = 2 * (f_prime_x**2 - f_x * f_double_prime_x)
        if denominator == 0:
            raise ValueError("Denominator in Halley's method is zero")

        x_next = x - (2 * f_x * f_prime_x) / denominator

        if abs(x_next - x) < tolerance:
            return x_next

        x = x_next

    raise ValueError("Halley's method did not converge")
def karatsuba(x, y):
    """Karatsuba multiplication algorithm."""
    if x < 10 or y < 10:
        return x * y

    m = min(len(str(x)), len(str(y)))
    m2 = m // 2

    # Split the digit sequences
    x1, x0 = divmod(x, 10**m2)
    y1, y0 = divmod(y, 10**m2)

    # 3 recursive calls
    z2 = karatsuba(x1, y1)
    z0 = karatsuba(x0, y0)
    z1 = karatsuba(x1 + x0, y1 + y0) - z2 - z0

    return z2 * 10**(2*m2) + z1 * 10**m2 + z0
def karatsuba_sqrt(n):
    """Compute the integer square root of n using Karatsuba multiplication."""
    if n < 0:
        raise ValueError("Cannot compute square root of a negative number")
    if n == 0:
        return 0

    # Initial guess (can be optimized)
    x = n

    while True:
        x_next = (x + n // x) // 2
        if x_next >= x:
            return x
        x = x_next

# Number to compute the square root of
number = 1234567890123451234567895678901234567890123456789056781111111111111187999999999334567

# Timing function
def time_method(method, *args):
    start_time = time.time()
    method(*args)
    return time.time() - start_time

# Measure execution times
time_newton = time_method(sqrt_newton, number)
time_halley = time_method(sqrt_halley, number)
time_karatsuba = time_method(karatsuba_sqrt, number)


print(f"Newton's Method: {time_newton:.6f} seconds")
print(f"Halley's Method: {time_halley:.6f} seconds")
print(f"Karatsuba Square Root: {time_karatsuba:.6f} seconds")
```
